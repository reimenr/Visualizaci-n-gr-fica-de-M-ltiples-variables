Extraer el año de la columna "Title" y agregarlo a la nueva columna "Year" y Eliminar el año de la columna "Title"

```python
movies["Year"] = movies["Title"].str.extract(r"\((\d{4})\)", expand=False)
movies["Title"] = movies["Title"].str.replace(r"\(\d{4}\)", "").str.strip()
movies.head(20)
```

Conservar solo el primer género en la columna "Genres"

```python
movies["Genres"] = movies["Genres"].str.split("|").str[0]
movies.head(20)
```

```python
movies1= movies.rename(columns={"Title": "Nombre de la película", "Year": "Año", "Genres": "Género"})
movies1.head(20)
```

Del ejercicio anterior me parece relevante cuantificar la cantidad de peliculas que se hicieron por genero, con esta informacion podriamos saber cual fue el genero del cual se produjeron mas peliculas en un lapso de tiempo de 90 años y la importancia para la inductria audiovisual de cada genero, que va ligado con el consumo de dicho tipo de genero por las personas. y tambien nos ayuda a saber cual es el genero del cual se generan menos peliculas y por tanto concluirimas que no es rentable hacer peliculas de este genero.

```python
movies1.groupby("Género").size()
```

```python
movies1.Género.value_counts().plot(kind="bar")
plt.title("Cantidad de peliculas por genero")
plt.show()
```

De la anterior grafica podemos concluir que el genero que mas produjo peliculas en un timepo de 90 años, fue Drama y comedia, por lo cual entendemos que fueron los generos mas rentables y de mayor consumo. mientras que se hicieron pocas peliculas de fantasia, para lo cual podemos pensar que no le intereso este genero tanto a las productoras como al consumidor. 

```python
movies1['Año'] = pd.to_numeric(movies1['Año'], errors='coerce')
```

```python
movies_año = movies1.groupby("Año")["Nombre de la película"].count()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(movies_año.index, movies_año.values, marker='o', linestyle='-', color='b')
plt.title('Cantidad de películas por año')
plt.xlabel('Año')
plt.ylabel('Cantidad de películas')
plt.grid(True)
plt.xticks(np.arange(min(movies_año.index), max(movies_año.index)+1, 5))
plt.show()
```

De la grafica anterior podemos concluir que el boom de produccion de peliculas de cualuqier genero se llevo a cabo entre los años 1990 y 1999 con mas de 150 peliculas cada año. cambiando totalmente la produccion de peliculas, que aunque desde 1919 empezo un aumento progresivo de produccion de peliculas, no fue hasta los años 80s que se empezaron a ver el mayor numero de producciones para luego en los 90s crecer muy rapido. 

sin embargo vemos que aunque en el año 1999 la produccion de peliculas fue alta respecto a 1980 hacia atras, se vio una reduccion de produccion de peliculas con fuertes caidas, incluso enc deajo de la media.
